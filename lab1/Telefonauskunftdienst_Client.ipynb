### Server Setup

## GETALL from server

```python
from pprint import pprint
import clientserver

client = clientserver.Client() # Create a client instance
response = client.call("GETALL", {"name": "John Doe"})

pprint(response)

client.close()
```

## GET specific entry from server

```python
from pprint import pprint
import clientserver

client = clientserver.Client() # Create a client instance

while True:
    try:
        name = input("Enter a name to look up (or 'exit' to quit): ")
        if name.lower() == 'exit':
            break
        response = client.call("GET", {"name": name})
        pprint(response)
    except KeyboardInterrupt:
        break

client.close()
```
